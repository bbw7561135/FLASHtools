```python
ts = []
Ekins, Eths, Etots, Emags = [], [], [], []

for ds in dss[:end]:
    print ds
    sp = ds.sphere(center, radius)
    t = ds.current_time
    ts.append(t)
    
    Eth = sum(sp['thermal_energy']*sp['cell_mass'])
    Ekin = sum(sp['kinetic_energy']/sp['density']*sp['cell_mass'])
    Emag = sum(sp['magnetic_pressure']*sp['cell_volume'])
    Etot = Eth + Ekin
    Eths.append(Eth)
    Ekins.append(Ekin)
    Etots.append(Etot)
    Emags.append(Emag)
```

```python
plt.plot(ts, Etots, label='total E')
plt.plot(ts, Ekins, label='kinetic E')
plt.plot(ts, Eths, label='thermal E')
plt.plot(ts, Emags, label='magnetic E')
plt.legend(loc=0)
```
