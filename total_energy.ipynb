```python
ts = []
Ekins, Eths, Etots = [], [], []

for ds in dss[:end]:
    print ds
    sp = ds.sphere(center, radius)
    t = ds.current_time
    ts.append(t)
    
    Eth = sum(sp['thermal_energy']*sp['cell_mass'])
    Ekin = sum(sp['kinetic_energy']/sp['density']*sp['cell_mass'])
    Etot = Eth + Ekin
    Eths.append(Eth)
    Ekins.append(Ekin)
    Etots.append(Etot)
```
