```python
#%pdb
import numpy as np
import time
import matplotlib.pyplot as plt
from scipy.special import kv
from scipy.integrate import quad
from scipy.optimize import curve_fit
matplotlib.rcParams['savefig.dpi'] = 150
```

```python
me = 9.109E-28
c = 2.998E10
e = 4.803E-10
B53 = lambda x: kv(5./3., x)

def F(x):
    x = min(x, 1E3)
    integral, err = quad(B53, x, np.Inf)#, epsabs=0, epsrel=1.49e-08)
    #print x, np.Inf, integral
    ub = 1E9*x
    while integral < 0.0 or np.isnan(integral):
        integral, err = quad(B53, x, ub)
        #print x, ub, integral
        ub /= 10
        if ub <= x: raise ArithmeticError
    return x*integral

def F_fit(x, A, pind):
    return A*x**pind*np.exp(-x)
```

```python
xx=np.logspace(-6, np.log10(4), 1000)
Fxx = np.array([F(x) for x in xx])
plt.plot(xx,Fxx)
plt.xlabel('x')
plt.ylabel('F(x)')
```

```python
xx=np.logspace(-6,2, 1000)
Fxx = np.array([F(x) for x in xx])
par, cov = curve_fit(F_fit, xx, Fxx, p0=[2.0, 0.33])
A, pind = par
print par
print cov

def F_fit_log10(x, A, pind):
    return np.log10(A*x**pind*np.exp(-x))

par, cov = curve_fit(F_fit_log10, xx, np.log10(Fxx), p0=[2.0, 0.33])
A2, pind2 = par
print par
print cov
```

```python
# Integrate over gamma
def func(gamma, B, nu):
    #return 3**0.5*B*e**3/c**2/me**2*F(4*np.pi*me*c/3.0/gamma**2/e/B*nu)*gamma**(-2)
    return F(4*np.pi*me*c/3.0/gamma**2/e/B*nu)*gamma**(-2)

def J(gmax, gmin, B, nu):
    #print '(gmin,gmax) = ', gmin, gmax
    #print '(xmin,xmax) = ', 4*np.pi*me*c/3.0/gmax**2/e/B*nu, 4*np.pi*me*c/3.0/gmin**2/e/B*nu
    integral, err = quad(func, gmin, gmax, args=(B, nu))
    #print integral, err
    return 3**0.5*B*e**3/c**2/me**2*integral
```

```python
class Synchrotron():
    B = 0.0
    gmax = 0.0
    gmin = 0.0
    nus = 0.0
    Js = 0.0
    nucmin = 0.0
    nuc = 0.0
    
    def gen_sync(self, B, gmax, gmin=1.0, lognumin=-5, lognumax=13, n_nu=51):
        self.B = B
        self.gmax = gmax
        self.gmin = gmin
        self.nucmin = 3.0*gmin**2*e*B/(4.0*np.pi*me*c)
        self.nuc = 3.0*gmax**2*e*B/(4.0*np.pi*me*c)
        t0 = time.time()
        nus = np.logspace(lognumin, lognumax, n_nu)
        Js = np.array([J(gmax, gmin, B, nu) for nu in nus])
        self.nus = nus[Js.nonzero()]
        self.Js = Js[Js.nonzero()]
        print 'Synchrotron Spectra created with B = %.2e gauss, (gmax, gmin) = (%.1e, %.1e)' % (B, gmax, gmin)
        print 'Elapsed time:', time.time()-t0
    
    __init__ = gen_sync
    
    def f1(self, nus, A1, pind):
        return self.nucmin**(-5./6.)*A1*nus**pind
    
    def f2(self, nus, A2):
        return A2*nus**(-0.5)*np.exp(-nus/self.nuc)
    
    def f(self, nus, A1, pind1, A2):
        return 1./(1./self.f1(nus, A1, pind1) + 1./self.f2(nus, A2))
    
    def f_log(self, nus, A1, pind1, A2):
        return np.log10(1./(1./self.f1(nus, A1, pind1) + 1./self.f2(nus, A2)))
    
    par3 = [0.0, 1./3., 0.0]
    
    def f_log_Aonly(self, nus, A1, A2):
        return np.log10(1./(1./self.f1(nus, A1, self.par3[1]) + 1./self.f2(nus, A2)))
        
    par1 = [2E5, 1./3., 1.]
    par2 = [2E5, 1./3., 1.]
    def fit_sync(self):
        self.par1, cov1 = curve_fit(self.f, self.nus, self.Js, p0=[1, 1./3., 1.])
        #print par1
        #print cov1
        self.par2, cov2 = curve_fit(self.f_log, self.nus, np.log10(self.Js), p0=[1, 1./3., 1.])
        #print par2
        #print cov2
        (self.par3[0], self.par3[2]), cov3 = curve_fit(self.f_log_Aonly, self.nus, np.log10(self.Js), p0=[1, 1])
        #print par3
        #print cov3
    
    def plot_sync(self, show=True):
        nus = self.nus
        Js = self.Js
        par1 = self.par1
        par2 = self.par2
        par3 = self.par3
        f = self.f
        fig = plt.figure(figsize=(8,6))
        ax1=fig.add_axes((.1,.4,.8,.5))
        ax1.plot(nus[::2], Js[::2], 'x', label='J(x)', c='k', ms=4)
        #ax1.plot(nus, f1(nus, par1[0], par1[1]))
        #ax1.plot(nus, f2(nus, par1[2], par1[3]))
        label = r'$(\frac{1}{%.3E\nu_{cmin}^{-5/6}\nu^{%.3f}}+\frac{1}{%.3E\nu^{-0.5}e^{-x}})^{-1}$' 
        ax1.plot(nus, f(nus, par1[0], par1[1], par1[2]), c='g', ls=':', alpha=0.5,
                 label=label % tuple(par1) +' (linear fit)')
        ax1.plot(nus, f(nus, par2[0], par2[1], par2[2]), c='b', ls='--', alpha=0.5,
                 label=label % tuple(par2) +' (log fit)')
        ax1.plot(nus, f(nus, par3[0], par3[1], par3[2]), c='r', ls='-.', alpha=0.5,
                 label=label % tuple(par3) +' (log fit)')
        ax1.vlines(self.nucmin, 1E-7, 1E1, lw=5, colors='grey', alpha=0.2)
        ax1.vlines(self.nuc, 1E-7, 1E1, lw=5, colors='grey', alpha=0.2)
        ax1.semilogx()
        ax1.semilogy()
        ax1.set_xlim(1E-5, 1E13)
        ax1.set_ylim(1E-7, 1E1)
        ax1.legend(loc=3)
        txt = ('B = %.0E gauss\n'+r'$\gamma_{max}$ = %.0E'+'\n'+r'$\gamma_{min}$ = %.0E') % (self.B, self.gmax, self.gmin)
        ax1.annotate(txt, (0.8,0.8), xycoords='axes fraction')
        ax1.xaxis.set_ticks_position('top')
        ax1.grid()
        
        ax2=fig.add_axes((.1,.25,.8,.15))
        ax2.plot(nus, f(nus, par1[0], par1[1], par1[2])-Js, c='g', ls=':')
        ax2.plot(nus, f(nus, par2[0], par2[1], par2[2])-Js, c='b', ls='--')
        ax2.plot(nus, f(nus, par3[0], par3[1], par3[2])-Js, c='r', ls='-.')
        ax2.semilogx()
        ax2.set_xlim(1E-5, 1E13)
        ax2.set_ylabel('residue')
        ax2.set_ylim(-0.5, 0.5)
        ax2.set_yticks([-0.4, -0.2, 0.00, 0.2, 0.4])
        ax2.set_yticklabels([-0.4, -0.2, 0.00, 0.2])
        ax2.hlines(0, nus.min(), nus.max(), lw=0.2, alpha=0.5)
        ax2.set_xticklabels([])
        
        ax3=fig.add_axes((.1,.1,.8,.15))
        ax3.plot(nus, (f(nus, par1[0], par1[1], par1[2])-Js)/Js*100, c='g', ls=':')
        ax3.plot(nus, (f(nus, par2[0], par2[1], par2[2])-Js)/Js*100, c='b', ls='--')
        ax3.plot(nus, (f(nus, par3[0], par3[1], par3[2])-Js)/Js*100, c='r', ls='-.')
        #plt.semilogy()
        ax3.semilogx()
        ax3.set_xlim(1E-5, 1E13)
        ax3.set_xlabel(r'$\nu$ (Hz)')
        ax3.set_ylabel('% error')
        ax3.set_ylim(-100,100)
        ax3.set_yticks([-100, -50, 0, 50, 100])
        ax3.set_yticklabels([-100, -50, 0, 50])
        ax3.hlines(0, nus.min(), nus.max(), lw=0.2, alpha=0.5)
        plt.savefig('./synchrotron_spectra_fitting/B%i_gmax%i_gmin%i.png' % (np.log10(self.B), np.log10(self.gmax), np.log10(self.gmin)))
        if show:
            plt.show()
        return fig
```

```python
for B in [1E-5, 1E-6]:
    for gmax in [1E3, 1E4, 1E5]:
        for gmin in [1E0, 1E1]:
            sync = Synchrotron(B, gmax, gmin)
            sync.fit_sync()
            sync.plot_sync(show=False)
```
