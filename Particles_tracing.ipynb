```python
import numpy as np
import util
import multiprocessing
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 150
import matplotlib.pyplot as plt
import yt
import logging
logging.getLogger('yt').setLevel(logging.ERROR)
pool = None

class Particle():
    def __init__(self, tag, maxInd):
        self.tag = tag
        self.t = np.zeros(maxInd)
        self.posx = np.zeros(maxInd)
        self.posy = np.zeros(maxInd)
        self.posz = np.zeros(maxInd)
        self.dens = np.zeros(maxInd)
        self.gamc = np.zeros(maxInd)
        self.B = np.zeros(maxInd)

    def read_at_time(self, ind, particles_at_time):
        if self.tag < 0:
        #    self.tag = tag
        #else if self.tag != tag:
            raise Exception
        
        if self.tag in particles_at_time.tag:
            part_ind = abs(particles_at_time.tag-self.tag).argmin()
            self.t[ind] = particles_at_time.time
            self.posx[ind] = particles_at_time.posx[part_ind]
            self.posy[ind] = particles_at_time.posy[part_ind]
            self.posz[ind] = particles_at_time.posz[part_ind]
            self.dens[ind] = particles_at_time.dens[part_ind]
            self.gamc[ind] = particles_at_time.gamc[part_ind]
            self.B[ind] = particles_at_time.B[part_ind]
        

class Particles_at_time():
    def __init__(self, time, all_data):
        self.time = time
        self.tag = all_data['all', 'particle_tag']
        self.posx = all_data['all', 'particle_position_x']
        self.posy = all_data['all', 'particle_position_y']
        self.posz = all_data['all', 'particle_position_z']
        self.dens = all_data['all', 'particle_dens']
        self.gamc = all_data['all', 'particle_gamc']
        self.B = np.sqrt(all_data['all', 'particle_magx']**2 \
                         + all_data['all', 'particle_magy']**2 \
                         + all_data['all', 'particle_magz']**2)*np.sqrt(4.0*np.pi)

class Sim():
    def __init__(self, plotfiles, partfiles):
        self.plotfiles = plotfiles
        self.partfiles = partfiles

        self.particles = []
        #self.t = multiprocessing.Array('d', len(plotfiles))
        self.t = np.zeros(len(plotfiles))
        ds = yt.load(self.plotfiles[-1].fullpath, particle_filename=self.partfiles[-1].fullpath)      
        for tag in sorted(ds.all_data()['all', 'particle_tag']):
            self.particles.append(Particle(tag, len(plotfiles)))


    def readone(self,i):
        #try:
        #print self.plotfiles[i].fullpath
        ds = yt.load(self.plotfiles[i].fullpath, particle_filename=self.partfiles[i].fullpath)
        ad = ds.all_data()
        #self.t[i] = ds.current_time
        return ds.current_time, Particles_at_time(ds.current_time, ad)
        #except:
        #    print 'KeyboardInterrupt catched...'
        #    raise Exception

        
    def readall(self):
        #nProcessor = multiprocessing.cpu_count()
        #pool = multiprocessing.Pool(nProcessor)
        
        for i in range(len(self.plotfiles)):
            time, pt_at_time = self.readone(i)
            self.t[i] = time
            for j in range(len(self.particles)):
                self.particles[j].read_at_time(i, pt_at_time)
        #results = pool.map(self.readone, range(len(plotfiles)))
        
        #for ii, tt, pt in results:
        #    self.t[ii] = tt
        #    self.particle[ii] = pt
            
        #pool.close()
        #pool.join()
```
